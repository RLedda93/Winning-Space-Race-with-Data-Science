```python
# Task 2: Mark the success/failed launches for each site on the map
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]

# Group by Launch Site and take the first row for each site to get unique launch sites' coordinates
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]

# Create a folium Map object centered at NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Function to determine marker color based on launch success
def get_marker_color(success):
    return 'green' if success else 'red'

# Add markers for each launch site to the map
for index, site in launch_sites_df.iterrows():
    site_name = site['Launch Site']
    site_coords = (site['Lat'], site['Long'])
    
    # Determine the success/failure of launches at this site
    success = spacex_df[spacex_df['Launch Site'] == site_name]['class'].astype(bool).all()
    
    # Create a marker for the launch site
    marker_color = get_marker_color(success)
    folium.Circle(
        site_coords,
        radius=1000,
        color=marker_color,
        fill=True,
        fill_color=marker_color,
        fill_opacity=0.6,
        popup=f"{site_name}<br>{'Success' if success else 'Failure'}"
    ).add_to(site_map)

# Display the map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities

# Function to calculate distance between two points using Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance between two points
    on the Earth (specified in decimal degrees)
    """
    # Earth's radius in kilometers
    radius = 6371.0
    
    # Convert decimal degrees to radians
    lat1_rad, lon1_rad, lat2_rad, lon2_rad = map(pd.np.radians, [lat1, lon1, lat2, lon2])

    # Calculate differences in latitudes and longitudes
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad

    # Haversine formula
    a = pd.np.sin(dlat / 2)**2 + pd.np.cos(lat1_rad) * pd.np.cos(lat2_rad) * pd.np.sin(dlon / 2)**2
    c = 2 * pd.np.arcsin(pd.np.sqrt(a))

    # Calculate distance
    distance_km = radius * c
    return distance_km

# Example: Calculate distance between two launch sites
launch_sites_df = spacex_df.groupby('Launch Site').first().reset_index()

# Select two launch sites for distance calculation
site1 = launch_sites_df.loc[0]  # Example: First launch site
site2 = launch_sites_df.loc[1]  # Example: Second launch site

# Calculate distance using Haversine formula
distance_km = haversine(site1['Lat'], site1['Long'], site2['Lat'], site2['Long'])
distance_miles = distance_km * 0.621371  # Convert to miles

print(f"Distance between {site1['Launch Site']} and {site2['Launch Site']}: {distance_km:.2f} km ({distance_miles:.2f} miles)")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.plugins import MousePosition
from math import sin, cos, sqrt, atan2, radians
from folium.features import DivIcon

# Function to calculate distance between two points using Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance between two points
    on the Earth (specified in decimal degrees)
    """
    # approximate radius of Earth in km
    R = 6373.0

    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)

    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad

    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Example: Coordinates of CCAFS LC-40 (Launch Site)
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Example: Coordinates of the closest city (obtained from MousePosition)
city_lat = 28.5721
city_lon = -80.648

# Calculate distance between launch site and closest city
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_city_miles = distance_city * 0.621371  # Convert to miles

print(f"Distance to closest city: {distance_city:.2f} km ({distance_city_miles:.2f} miles)")

# Create a folium Map centered at NASA Johnson Space Center (or any desired location)
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)  # Adjust zoom_start as needed

# Add MousePosition to the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Add a Marker for the closest city point
city_marker = folium.Marker(
    [city_lat, city_lon],
    icon=folium.CustomIcon(icon_image='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_3/images/city.png',
                           icon_size=(20, 20)),
    popup=f"Distance: {distance_city:.2f} km"
)
site_map.add_child(city_marker)

# Create a PolyLine between launch site and closest city
lines_city = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [city_lat, city_lon]],
    weight=1,
    color='blue'
)
site_map.add_child(lines_city)

# Display the map
site_map
```
